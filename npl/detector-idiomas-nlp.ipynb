```python
def remover(textos, regex):
    if type(textos) == str:
        return regex.sub("", textos)
    else:
        return [regex.sub("", texto) for texto in textos]
```

```python
def substituir_codigo(textos, regex):
    if type(textos) == str:
        return regex.sub("CODE", textos)
    else:
        return [regex.sub("CODE", texto) for texto in textos]
```

```python
regex_espaco = re.compile(r"\s{2,}")
regex_quebra_linha = re.compile(r"(\n)")

def substituir_por_espaco(textos, regex):
    if type(textos) == str:
        return regex.sub(" ", textos)
    else:
        return [regex.sub(" ", texto) for texto in textos]
    
print(substituir_por_espaco("Alura        \n\n\n  Caelum", regex_espaco))
print(substituir_por_espaco("Alura        \n\n\n  Caelum", regex_quebra_linha))
```

```python
def trata_texto(dados: pd.DataFrame):
    questoes_sem_pont = remover(dados.sem_code_tag, regex_pontuacao)
    questoes_sem_pont_minus = minusculo(questoes_sem_pont)
    questoes_sem_pont_minus_dig = remover(questoes_sem_pont_minus, regex_digitos)
    questoes_sem_espaco_duplicado = remover(questoes_sem_pont_minus_dig, regex_espaco)
    questoes_sem_quebra_linha = remover(questoes_sem_espaco_duplicado, regex_quebra_linha)
    dados['questoes_tratadas'] = questoes_sem_quebra_linha
```

```python
from nltk.lm import MLE
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.tokenize import WhitespaceTokenizer
from sklearn.model_selection import train_test_split

def treinar_modelo_mle(lista_textos, n=2):
    todas_questoes = ' '.join(lista_textos)
    todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
    ngrams, vocab = padded_everygram_pipeline(n, todas_palavras)
    modelo = MLE(n)
    modelo.fit(ngrams, vocab)
    return modelo
```

```python
def calc_perplexidade(modelo, texto, n=2):
    perplexidade = 0
    palavras = WhitespaceTokenizer().tokenize(texto)
    palavras_fakechar = [list(pad_both_ends(palavra, n=n)) for palavra in palavras]
    palavras_ngrams = [list(bigrams(palavra)) for palavra in palavras_fakechar]
    for palavra in palavras_ngrams:
        perplexidade += modelo.perplexity(palavra)
    
    return perplexidade
```

```python
from nltk.lm import Laplace
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.tokenize import WhitespaceTokenizer
from sklearn.model_selection import train_test_split

def treinar_modelo_laplace(lista_textos, n=2):
    todas_questoes = ' '.join(lista_textos)
    todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
    ngrams, vocab = padded_everygram_pipeline(n, todas_palavras)
    modelo = Laplace(n)
    modelo.fit(ngrams, vocab)
    return modelo
```

```python
modelo_pt_lap = treinar_modelo_laplace(pt_treino)
modelo_es_lap = treinar_modelo_laplace(es_treino)
modelo_en_lap = treinar_modelo_laplace(en_treino)
```

```python
print(calc_perplexidade(modelo_pt_lap, pt_teste.iloc[1]))
print(calc_perplexidade(modelo_pt_lap, es_teste.iloc[1]))
print(calc_perplexidade(modelo_pt_lap, en_teste.iloc[1]))
```

```python
print(calc_perplexidade(modelo_es_lap, pt_teste.iloc[1]))
print(calc_perplexidade(modelo_es_lap, es_teste.iloc[1]))
print(calc_perplexidade(modelo_es_lap, en_teste.iloc[1]))
```

```python
print(calc_perplexidade(modelo_en_lap, pt_teste.iloc[1]))
print(calc_perplexidade(modelo_en_lap, es_teste.iloc[1]))
print(calc_perplexidade(modelo_en_lap, en_teste.iloc[1]))
```

```python
def atribui_idioma(lista_textos):
    idioma = []
    for texto in lista_textos:
        pt = calc_perplexidade(modelo_pt_lap, texto)
        es = calc_perplexidade(modelo_es_lap, texto)
        en = calc_perplexidade(modelo_en_lap, texto)
        if en >= pt <= es:
            idioma.append("pt")
        elif pt > en < es:
            idioma.append("en")
        else:
            idioma.append("es")
    return idioma

def taxa_acerto(lista_textos, idioma_alvo):
    resultados = atribui_idioma(lista_textos)
    return resultados.count(idioma_alvo)/len(resultados)
```

```python
print("Português:", taxa_acerto(pt_teste, "pt"))
```

```python
print("Espanhol:", taxa_acerto(es_teste, "es"))
```

```python
print("Inglês:", taxa_acerto(en_teste, "en"))
```
