## Observation Space

|Num|Observation|Min|Max|Unit|

## Action Space

There are 3 discrete deterministic actions:

* 0: Accelerate to the left
* 1: Donâ€™t accelerate
* 2: Accelerate to the right

```python
env = gym.make(env_name, render_mode="human")
observation = env.reset(seed=101)
for step in range(100):
    env.render()
    action = 2 # our cart goes right
    observation, reward, terminated, truncated, info = env.step(action)
    print(f"Position: {observation[0]}")
    print(f"Velocity: {observation[1]}")
    time.sleep(0.01)
env.close()
```

```python
def simple_agent(observation):
    # observation (partial state)
    position, velocity = observation
    if position > -0.1 and position < 0.4: # when to go right -->
        action = 2
    elif velocity < 0 and position < -0.2: # when to go left <--
        action = 0
    else: # when to do nothing
        action = 1
    return action
```

```python
env = gym.make(env_name, render_mode="human")
observation = env.reset(seed=42)
observation = observation[0]
for step in range(500):
    env.render()
    action = simple_agent(observation)
    observation, reward, terminated, truncated, info = env.step(action)
    if terminated or truncated:
        break
    time.sleep(0.001)
env.close()
```
